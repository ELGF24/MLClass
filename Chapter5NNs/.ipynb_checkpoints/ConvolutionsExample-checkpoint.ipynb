```python
from PIL import Image
import numpy as np
from scipy import signal as sg
import pylab as plt
%matplotlib inline 
```

```python
#we define some useful functions
def np_from_img(fname):
    #This transforms and image to a numpy array
    return np.asarray(Image.open(fname).convert('L'), dtype=np.float32) 
    #The .convert('L') is to transform color images to RGB

def save_as_img(ar, fname):
    #This saves a numpy array as an image
    Image.fromarray(ar.round().astype(np.uint8)).save(fname)

def norm(ar):
    #This normalizes images to be in a 0-255 range
    return 255.*np.absolute(ar)/np.max(ar)
```

```python
img = np_from_img('Enterprise.jpg')
img_1 = sg.convolve(img, [[1.],[-1.]])
img_2 = sg.convolve(img, [[1., -1.]])
```

```python
plt.imshow(img,cmap='Greys_r')
```

```python
plt.imshow(img_1, cmap = 'Greys_r')
```

```python
plt.imshow(img_2, cmap = 'Greys_r')
```
